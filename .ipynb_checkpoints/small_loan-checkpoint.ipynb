```python
import pandas as pd
small_loan = pd.read_csv('small_loan.csv')
small_loan.head(10)
```

```python
small_loan['response'].value_counts()
```

```python
small_loan['response'].value_counts(normalize=True)
```

```python
small_loan.replace({'YES':1, 'NO':0, 'M': 1, 'F': 0}, inplace=True)
small_loan = pd.get_dummies(small_loan, prefix=['region', 'children'], prefix_sep='_', columns=['region', 'children'])
```

```python
small_loan
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

X = small_loan.drop(['id','response'],axis=1)
y = small_loan['response']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=107)

#oversample = RandomOverSampler(sampling_strategy='minority', random_state=103)
#X_train,y_train = oversample.fit_resample(X_train,y_train)

rf = RandomForestClassifier(n_estimators=1000, max_depth=12, random_state=107)
rf.fit(X_train, y_train)
```

```python
preds = rf.predict(X_test)
print(accuracy_score(y_test,preds))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,preds))
```

```python
import joblib
joblib.dump(rf, 'small_loan_rf_model.pkl') 
```
